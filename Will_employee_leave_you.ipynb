```
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow import keras
from keras.metrics import *
from keras import backend
train = pd.read_csv('/content/Train.csv')

train['Gender'] = train['Gender'].replace('M', 0)
train['Gender'] = train['Gender'].replace('F', 1)
train['Gender'] = train['Gender'].replace(np.nan, 2)

train['Age'] = train['Age'].replace(np.nan, train['Age'].mean())
train['Age'] = train['Age']/train['Age'].mean()

train['Relationship_Status'] = train['Relationship_Status'].replace('Married', 0)
train['Relationship_Status'] = train['Relationship_Status'].replace('Single', 1)
train['Relationship_Status'] = train['Relationship_Status'].replace(np.nan, 2)

train['Hometown'] = train['Hometown'].replace('Clinton', 0)
train['Hometown'] = train['Hometown'].replace('Franklin', 1)
train['Hometown'] = train['Hometown'].replace('Lebanon', 2)
train['Hometown'] = train['Hometown'].replace('Springfield', 3)
train['Hometown'] = train['Hometown'].replace('Washington', 4)
train['Hometown'] = train['Hometown'].replace(np.nan, 5)

train['Unit'] = train['Unit'].replace('Accounting and Finance', 0)
train['Unit'] = train['Unit'].replace('Human Resource Management', 1)
train['Unit'] = train['Unit'].replace('IT', 2)
train['Unit'] = train['Unit'].replace('Logistics', 3)
train['Unit'] = train['Unit'].replace('Marketing', 4)
train['Unit'] = train['Unit'].replace('Operarions', 5)
train['Unit'] = train['Unit'].replace('Production', 6)
train['Unit'] = train['Unit'].replace('Purchasing', 7)
train['Unit'] = train['Unit'].replace('Quality', 8)
train['Unit'] = train['Unit'].replace('R&D', 9)
train['Unit'] = train['Unit'].replace('Sales', 10)
train['Unit'] = train['Unit'].replace('Security', 11)
train['Unit'] = train['Unit'].replace(np.nan, 12)

train['Decision_skill_possess'] = train['Decision_skill_possess'].replace('Analytical', 0)
train['Decision_skill_possess'] = train['Decision_skill_possess'].replace('Behavioral', 1)
train['Decision_skill_possess'] = train['Decision_skill_possess'].replace('Conceptual', 2)
train['Decision_skill_possess'] = train['Decision_skill_possess'].replace('Directive', 3)
train['Decision_skill_possess'] = train['Decision_skill_possess'].replace(np.nan, 4)

train['Compensation_and_Benefits'] = train['Compensation_and_Benefits'].replace('type0', 0)
train['Compensation_and_Benefits'] = train['Compensation_and_Benefits'].replace('type1', 1)
train['Compensation_and_Benefits'] = train['Compensation_and_Benefits'].replace('type2', 2)
train['Compensation_and_Benefits'] = train['Compensation_and_Benefits'].replace('type3', 3)
train['Compensation_and_Benefits'] = train['Compensation_and_Benefits'].replace('type4', 4)
train['Compensation_and_Benefits'] = train['Compensation_and_Benefits'].replace(np.nan, 5)

train['Education_Level'] = train['Education_Level'].replace(np.nan, train['Education_Level'].mean())
train['Time_of_service'] = train['Time_of_service'].replace(np.nan, train['Time_of_service'].mean())
train['Time_since_promotion'] = train['Time_since_promotion'].replace(np.nan, train['Time_since_promotion'].mean())
train['growth_rate'] = train['growth_rate'].replace(np.nan, train['growth_rate'].mean())
train['Travel_Rate'] = train['Travel_Rate'].replace(np.nan, train['Travel_Rate'].mean())
train['Post_Level'] = train['Post_Level'].replace(np.nan, train['Post_Level'].mean())
train['Pay_Scale'] = train['Pay_Scale'].replace(np.nan, train['Pay_Scale'].mean())
train['Work_Life_balance'] = train['Work_Life_balance'].replace(np.nan, train['Work_Life_balance'].mean())
train['VAR1'] = train['VAR1'].replace(np.nan, train['VAR1'].mean())
train['VAR2'] = train['VAR2'].replace(np.nan, train['VAR2'].mean())
train['VAR3'] = train['VAR3'].replace(np.nan, train['VAR3'].mean())
train['VAR4'] = train['VAR4'].replace(np.nan, train['VAR4'].mean())
train['VAR5'] = train['VAR5'].replace(np.nan, train['VAR5'].mean())
train['VAR6'] = train['VAR6'].replace(np.nan, train['VAR6'].mean())
train['VAR7'] = train['VAR7'].replace(np.nan, train['VAR7'].mean())
train = train.replace(np.nan, 0)

train_x = train[['Gender','Age','Education_Level','Relationship_Status','Hometown','Unit','Decision_skill_possess','Time_of_service','Time_since_promotion','growth_rate','Travel_Rate','Post_Level','Pay_Scale','Compensation_and_Benefits','Work_Life_balance','VAR1','VAR2','VAR3','VAR4','VAR5','VAR6','VAR7']]
train_y = train[['Attrition_rate']]

train_x = np.array(train_x)
train_y = np.array(train_y)
train
```

```
np.shape(train_y)
```

```
model = tf.keras.Sequential([keras.layers.Dense(22, input_dim = 22, activation = tf.nn.leaky_relu),
                             tf.keras.layers.Dense(11, activation = tf.nn.leaky_relu),
                             tf.keras.layers.Dense(1, activation = tf.nn.sigmoid)]
                            )
# def rmse(y_true, y_pred):
# 	return backend.sqrt(backend.mean(backend.square(y_pred - y_true), axis = -1))
model.compile(
    optimizer = tf.keras.optimizers.RMSprop(),
    loss = 'mse',
    metrics=tf.keras.metrics.RootMeanSquaredError()
    )
model.fit(train_x, train_y, validation_split=0.2,epochs = 25,batch_size = 16)
```
